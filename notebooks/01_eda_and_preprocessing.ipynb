```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

fraud_df = pd.read_csv('../data/raw/Fraud_Data.csv')
ip_df = pd.read_csv('../data/raw/IpAddress_to_Country.csv')

print(fraud_df.shape)
print(ip_df.shape)
```

```python
print('___missing values in Fraud Data___')
print(fraud_df.isnull().sum())

print('__missing values in IP data__')
print(ip_df.isnull().sum())

print('---data types---')
print(fraud_df.dtypes)
```

```python
# cast objects to datetimes
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])
# calculat the differnce in secondss

fraud_df['time_diff'] = (fraud_df['purchase_time'] - fraud_df['signup_time']).dt.total_seconds()

print(fraud_df[['signup_time', 'purchase_time', 'time_diff']].head())
```

```python
# 1. Sort both dataframes by the IP columns (Crucial for merge_asof!)
fraud_df = fraud_df.sort_values('ip_address')
ip_df = ip_df.sort_values('lower_bound_ip_address')

# 2. Use merge_asof to find the country
# This looks for the 'lower_bound' that is closest to our 'ip_address'
fraud_with_country = pd.merge_asof(
    fraud_df, 
    ip_df, 
    left_on='ip_address', 
    right_on='lower_bound_ip_address'
)

# 3. Clean up: Only keep the country if it actually falls within the upper bound
import numpy as np
fraud_with_country['country'] = np.where(
    fraud_with_country['ip_address'] <= fraud_with_country['upper_bound_ip_address'], 
    fraud_with_country['country'], 
    'Unknown'
)
```

```python
fraud_counts = fraud_with_country.groupby('country')['class'].sum().sort_values(ascending=False)

print('--- Top 10 countries by Fraud Volume ---')
print(fraud_counts.head(10))

# to see which source brings in the most fraud 
source_fraud = fraud_with_country.groupby('source')['class'].mean()
print("\n--- Fraud Rate by Marketing Source ---")
print(source_fraud)
```

```python

country_fraud_rate = fraud_with_country.groupby('country')['class'].mean().sort_values(ascending=False)

print("--- Top 10 Countries by Fraud RATE (Percentage) ---")
print(country_fraud_rate.head(10))
```

```python


# Create two groups: Fraud vs Legitimate
fraud = fraud_with_country[fraud_with_country['class'] == 1]
legit = fraud_with_country[fraud_with_country['class'] == 0]

# Plot the distribution of 'time_diff'
plt.figure(figsize=(10,6))
sns.histplot(fraud['time_diff'], label='Fraud', color='red', kde=True, bins=50)
sns.histplot(legit['time_diff'], label='Legit', color='blue', kde=True, bins=50)
plt.legend()
plt.title('Time Difference: Signup vs Purchase')
plt.show()
```

```python
# Count how many 0s and 1s we have
class_counts = fraud_with_country['class'].value_counts()
print("--- Class Distribution ---")
print(class_counts)

# Calculate the percentage
print("\n--- Percentage of Fraud ---")
print(class_counts / len(fraud_with_country) * 100)

# Visualize it
class_counts.plot(kind='bar', color=['blue', 'red'])
plt.title('Fraud vs Legitimate Count')
plt.show()
```
